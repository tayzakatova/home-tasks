```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('iris_data.csv')
#print(df)

a = []

for i in set(list(df['Species'])):
    a.append(list(df['Species']).count(i))


fig_1 = plt.figure(figsize = (18, 6))
plt.pie(a, labels = set(list(df['Species'])), autopct='%1.1f%%')
plt.title("  ")

#autopct='%1.1f%%'    

mic = df[df['PetalLengthCm'] <= 1.2].shape[0]
norm = df[(df['PetalLengthCm'] > 1.2) & (df['PetalLengthCm'] < 1.5)].shape[0]
big = df[df['PetalLengthCm'] >= 1.5].shape[0]

fig_2 = plt.figure(figsize = (18, 6))
plt.pie([mic, norm, big], labels = [' 1.2', ' 1.2   1.5', ' 1.5'], autopct='%1.1f%%')
plt.title("  ")



plt.show()
```
