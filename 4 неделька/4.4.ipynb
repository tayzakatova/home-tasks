```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('iris_data.csv')
a = [('SepalLengthCm', 'SepalWidthCm'), ('SepalLengthCm', 'PetalLengthCm'), ('SepalLengthCm', 'PetalWidthCm'), ('SepalWidthCm', 'PetalLengthCm'), ('SepalWidthCm', 'PetalWidthCm'), ('PetalLengthCm', 'PetalWidthCm')]


for x, y in a:
  plt.figure(figsize=(10, 10))
  plt.scatter(df[x], df[y], marker="1", color='green', label='Исходные данные')
  c = np.polyfit(df[x], df[y], 1)
  p = np.poly1d(c)
  plt.plot(df[x], p(df[x]), color='r', label='Прямая МНК')
  plt.xlabel(x)
  plt.ylabel(y)
  plt.title(f'{x} от {y}')
  plt.legend()


'''X = df[[x]].values
y = df[y].values
model = LinearRegression()
model.fit(X, y)'''
#Я очень пыталась вывести коэффициенты прямых мнк, но  меня ничего не вышло Т_Т


  #k = LinearRegression()
#print(f'Коэффициенты для {y} от {x}:', model.coef_[0], model.intercept_)
plt.show()
```
