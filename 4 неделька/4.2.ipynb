```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
pos = 0 #центр графика
scale = 10 #более плавный и разброс больше
a = np.random.normal(pos, 18, 1000)
b = np.random.normal(pos, 18, 5000)
c = np.random.normal(pos, 18, 10000)
d = np.random.normal(pos, 18, 20000)

fig = plt.figure(figsize = (18, 6))
x_1 = fig.add_subplot(221)
x_2 = fig.add_subplot(222)
x_3 = fig.add_subplot(223)
x_4 = fig.add_subplot(224)

x_1.hist(a, 100)
x_2.hist(b, 100)
x_3.hist(c, 100)
x_4.hist(d, 100)

plt.show
```
